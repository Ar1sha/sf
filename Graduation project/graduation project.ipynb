# Описание датасета

Period_Log.csv:
* user_id - ID пользователя 

* cycle_number - номер цикла(порядковый номер)

* start_date - Дата начала цикла

* cycle_length_days - Длительность текущего менструального цикла(в днях)(от начала одного цикла до начала следующего)

* prev_cycle_length - Длительность предыдущего цикла(в днях)

* cycle_phase - Фаза цикла:
    - Luteal - Лютеиновая фаза(после овуляции)
    - Follicular - фолликуоярная фаза(до овуляции)
    - Menstrual - менструальная фаза

* flow_level - Интенсивность выделений: 
    - Heavy- Сильные
    - Light - Слабые
    - Moderate - Средние

* pain_level(0–10) - Уровень боли(0-10)

* pms_symptoms(Yes/No) - Симптомы ПМС(Да/Нет)

* mood_score - Уровень настроения 

* energy_level - Уровень энергии 

* sleep_hours_cycle - Продолжительность сна в часах во время цикла

* concentration_score - Уровень концентрации 

* stress_score_cycle - Уровень стресса во время цикла

* work_hours_lost - Пропущенные часы из-за плохого самочувствия

* estrogen_pgml - Уровень эстрогена(pg/ml)

* progesterone_ngml - Уровень прогестерона(ng/ml)

* ovulation_result(Positive/Negative) - Результат овуляции(Положительный/Отрицательный)

* overall_health_score - Оценка общего самочувствия

* log_consistency_score -  Оценка регулярности ведения записей 

* prepared_before_period(0/1) - Подготовка перед периодом(0=нет/1=да)

User_Profile.csv

* user_id - ID пользователя

* state - штат проживания

* age - возраст

* bmi - индекс массы тела

* diet_quality - Качество диеты:
    - Good - хорошая
    - Poor - плохая
    - Excellent - идеальная
    - Average - средняя

* exercise_frequency - частота физических упражнений

* sleep_hours - средняя продолжительность сна(в часах)

* caffeine_intake - потребление кофеина

* water_intake_liters - потребление воды(в литрах)

* alcohol_consumption - частота питья алкоголя:
    - Occasionally - иногда
    - Weekly - еженедельно 
    - Never - никогда

* smoking_status - статус курения (no-не курит/yes-курит)

* birth_control_use - использование контрацепции(0-нет/1-да)

* pcos_diagnosed - Диагностирован ли диагностирован ли синдром поликистозных яичников(0-нет/1-да)

* stress_score_baseline - базовый уровень стресса
