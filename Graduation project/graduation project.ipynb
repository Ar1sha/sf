# Описание датасета

Period_Log.csv:
* user_id - ID пользователя 

* cycle_number - номер цикла(порядковый номер)

* start_date - Дата начала цикла

* cycle_length_days - Длительность текущего менструального цикла(в днях)(от начала одного цикла до начала следующего)

* prev_cycle_length - Длительность предыдущего цикла(в днях)

* cycle_phase - Фаза цикла:
    - Luteal - Лютеиновая фаза(после овуляции)
    - Follicular - фолликуоярная фаза(до овуляции)
    - Menstrual - менструальная фаза

* flow_level - Интенсивность выделений: 
    - Heavy- Сильные
    - Light - Слабые
    - Moderate - Средние

* pain_level(0–10) - Уровень боли(0-10)

* pms_symptoms(Yes/No) - Симптомы ПМС(Да/Нет)

* mood_score - Уровень настроения 

* energy_level - Уровень энергии 

* sleep_hours_cycle - Продолжительность сна в часах во время цикла

* concentration_score - Уровень концентрации 

* stress_score_cycle - Уровень стресса во время цикла

* work_hours_lost - Пропущенные часы из-за плохого самочувствия

* estrogen_pgml - Уровень эстрогена(pg/ml)

* progesterone_ngml - Уровень прогестерона(ng/ml)

* ovulation_result(Positive/Negative) - Результат овуляции(Положительный/Отрицательный)

* overall_health_score - Оценка общего самочувствия

* log_consistency_score -  Оценка регулярности ведения записей 

* prepared_before_period(0/1) - Подготовка перед периодом(0=нет/1=да)

User_Profile.csv

* user_id - ID пользователя

* state - штат проживания

* age - возраст

* bmi - индекс массы тела

* diet_quality - Качество диеты:
    - Good - хорошая
    - Poor - плохая
    - Excellent - идеальная
    - Average - средняя

* exercise_frequency - частота физических упражнений

* sleep_hours - средняя продолжительность сна(в часах)

* caffeine_intake - потребление кофеина

* water_intake_liters - потребление воды(в литрах)

* alcohol_consumption - частота питья алкоголя:
    - Occasionally - иногда
    - Weekly - еженедельно 
    - Never - никогда

* smoking_status - статус курения (no-не курит/yes-курит)

* birth_control_use - использование контрацепции(0-нет/1-да)

* pcos_diagnosed - Диагностирован ли диагностирован ли синдром поликистозных яичников(0-нет/1-да)

* stress_score_baseline - базовый уровень стресса


```python
# Импорт библиотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
# Загрузка данных 
df_period = pd.read_csv("csv/Period_Log.csv")
df_user = pd.read_csv("csv/User_Profile.csv")
```

```python
# Объединяем столбцы по user_id
df = df_period.merge(df_user, on='user_id', how='left')
df.head()
```

```python
# Переводим дату к datetime
df['start_date']=pd.to_datetime(df['start_date'])
```

```python
# Смотрим на пропуски
px.imshow(df.isna(), title='Тепловая карта пропусков')
```

```python
# Cмотрим процент пропусков в prev_cycle_length, exercise_frequency 
print(f"Процент пропусков в стобце prev_cycle_length: {round(df_period['prev_cycle_length'].isna().mean() * 100, 2)}%")
print(f"Процент пропусков в exercise_frequency: {round(df['exercise_frequency'].isna().mean() * 100,2)}%")
```

```python
# Преобразуем частоту физических упражнений(для статистики оставляем NaN, в дальнейшем при обучении модели мы их заменим/удалим/будем использовать модели, которые работают с NaN)
df['exercise_frequency'] = df['exercise_frequency'].str.replace('days/week', '', regex=False).str.strip()
```

```python
# Пропусков не так много, можно заменить на среднее
df['prev_cycle_length']=df['prev_cycle_length'].fillna(df['prev_cycle_length'].mean())
print(f"Пропуски после замены: {df['prev_cycle_length'].isna().sum()} ")
```

```python
# Создаем отдельный дата сет для корреляции
df_corr = df.drop(['user_id', 'state', 'start_date', 'activity_score'], axis=1)
# Заменяем категориальные строки числами
df_corr['cycle_phase'] = df['cycle_phase'].replace({
    'Luteal': 0, 
    'Follicular': 1,
    'Menstrual': 2
})
df_corr['flow_level'] = df['flow_level'].replace({
    'Light':0,
    'Moderate':1,
    'Heavy': 2
})
df_corr['pms_symptoms'] = df['pms_symptoms'].replace({
    'Yes':1,
    'No':0
})
df_corr['ovulation_result'] = df['ovulation_result'].replace({
    'Positive':1,
    'Negative':0
})
df_corr['diet_quality'] = df['diet_quality'].replace({
    'Excellent':4,
    'Good':3,
    'Average':2,
    'Poor':1
})
df_corr['exercise_frequency'] = df_corr['exercise_frequency'].replace({
    '5–6': 3,
    '3–4':2,
    '1–2':1
})
df_corr['alcohol_consumption'] = df_corr['alcohol_consumption'].replace({
    'Occasionally':2,
    'Weekly':3,
    'Never':1
})
df_corr['smoking_status'] = df['smoking_status'].replace({
    'Yes':1,
    'No':0
})

df_corr = df_corr.dropna()
# Линейная зависимость признаков
df_corr = df_corr.corr()
# Отображение линейной зависимости

px.imshow(df_corr, title='Корреляция признаков')
```
